### Conditional statements:
- Conditional statements helps you to take decisions
- IF, ELSE, ELIF are the examples of conditional statement in python

```python
age = 25

#giving conditional
if age >=18:
    print("Congrats!! you are eligible to vote")

else:
    print("oohh sorry!! come next year")
```

```python
marks = 75

if marks >=90:
    print('Outstanding')
elif marks >=60:
    print('Good')
else:
    print("Try again!!")
```

### Loops:
- Loops are used when you want to do tasks repeatedly.
- It reduces time  and repeating process.
- There are 2 loops : FOR, WHILE

##### We use FOR LOOP when we want to go through items one by one:

```python
numbers =[10,20,30]

for num in numbers:
    print(num)
```

##### Use WHILE LOOPS when you want to run until the condition satisfies:

```python
count = 1
while count <= 3:
    print(count)
    count += 1
```

- Keep running this block of code as long as a condition is true
- CONDITION is: keep looping while count is less than or equal to 3.

#### Sort Descending
- Using reverse =True

```python
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort(reverse = True)
print(thislist)
```

##### Case Insensitive Sort

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort()
print(thislist)
```

##### Case sensitive Sort

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort(key = str.lower)
print(thislist)
```

##### Reverse

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.reverse()
print(thislist)
```

## List Mehtods:

| Method    | Description                                                                 |
|-----------|-----------------------------------------------------------------------------|
| append()  | Adds an element at the end of the list                                      |
| clear()   | Removes all the elements from the list                                      |
| copy()    | Returns a copy of the list                                                  |
| count()   | Returns the number of elements with the specified value                     |
| extend()  | Adds elements of another list (or iterable) to the end of the current list  |
| index()   | Returns the index of the first element with the specified value             |
| insert()  | Adds an element at a specified position                                     |
| pop()     | Removes the element at a specified position                                 |
| remove()  | Removes the item with the specified value                                   |
| reverse() | Reverses the order of the list                                              |
| sort()    | Sorts the list                                                              |
