### Dictionary Frequency Counter

```python
text = "apple banana apple"
words=text.split()

freq={}

for word in words:
    if word in freq:
        freq[word] +=1

    else:
        freq[word]=1

print(freq)
```

### Prime Number Check

```python
num = 7

is_prime = True

for i in range(2, num):
    if num % i == 0:
        is_prime = False

print(is_prime)
```

### Count Character Frequency

```python
text = "hello"

freq = {}

for char in text:
    freq[char] = freq.get(char, 0) + 1

print(freq)
```

### Sort List of Dictionaries

```python
students = [
    {"name": "A", "age": 24},
    {"name": "B", "age": 20}
]

sorted_students = sorted(students,key=lambda x:x['age'])
sorted_students
```

### Filter Data

```python
numbers = [1,2,3,4,5,6]

even = list(filter(lambda x: x % 2 == 0, numbers))

print(even)
```

### Map Function

```python
numbers = [1,2,3]

squared = list(map(lambda x: x*x, numbers))

print(squared)
```
