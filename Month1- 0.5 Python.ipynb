## Create an Iterator
We want to create something that:

1 2 3 4 5 ...

one number at a time.

Like a machine that gives the “next number” whenever asked.

First important idea:

- Python loops work using:

next()

```python
nums = [1, 2, 3]

it = iter(nums)

print(next(it))
```

Python asks:

“What is the next value?”

#### Now we create OUR OWN iterator
Step 1: Create class

Step 2: __iter__()

```python
class mynumber:
    def __iter__(self):
        self.a =1 ##“Start counting from 1”
        return self
```

Step 3: __next__()

```python
def __next__(self):
    x = self.a
    self.a += 1
    return x
```

#### Parse JSON - Convert from JSON to Python

```python
import json

# some JSON:
x =  '{ "name":"John", "age":30, "city":"New York"}'

y=json.loads(x)
y
```

#### Convert from Python to JSON
If you have a Python object, you can convert it into a JSON string by using the json.dumps() method.

```python
# some JSON:
a =  { "name":"John", "age":30, "city":"New York"}
b= json.dumps(a)

print(b)
```

#### You can convert Python objects of the following types, into JSON strings:

- dict
- list
- tuple
- string
- int
- float
- True
- False
- None

```python
import json

print(json.dumps({"name": "John", "age": 30}))
print(json.dumps(["apple", "bananas"]))
print(json.dumps(("apple", "bananas")))
print(json.dumps("hello"))
print(json.dumps(42))
print(json.dumps(31.76))
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None))
```

```python
import json

x = {
  "name": "John",
  "age": 30,
  "married": True,
  "divorced": False,
  "children": ("Ann","Billy"),
  "pets": None,
  "cars": [
    {"model": "BMW 230", "mpg": 27.5},
    {"model": "Ford Edge", "mpg": 24.1}
  ]
}

print(json.dumps(x))
```

```python
json.dumps(x, indent=4)
```

```python
json.dumps(x, indent=4, separators=(". ", " = "))
```

```python
json.dumps(x, indent=4, sort_keys=True)
```

## RegEx
A RegEx, or Regular Expression, is a sequence of characters that forms a search pattern.

RegEx can be used to check if a string contains the specified search pattern.

- Python has a built-in package called re, which can be used to work with Regular Expressions.

Import the re module:

```python
import re
```

Very important in:

data cleaning

NLP

web scraping

log analysis

preprocessing pipelines

```python
txt = "The rain in Spain"
x= re.search("^The.*Spain$",txt)

if x:
    print("Yes! we have a match")
else:
    print('not found')
```

### RegEx Functions

| Function    | Description                                                       |
| ----------- | ----------------------------------------------------------------- |
| `findall()` | Returns a list containing all matches                             |
| `search()`  | Returns a Match object if there is a match anywhere in the string |
| `split()`   | Returns a list where the string has been split at each match      |
| `sub()`     | Replaces one or many matches with a string                        |



## Metacharacters

| Character | Description                         | Example   |        |        |
| --------- | ----------------------------------- | --------- | ------ | ------ |
| `[]`      | Set of characters                   | `[a-m]`   |        |        |
| `\`       | Special sequence / escape character | `\d`      |        |        |
| `.`       | Any character except newline        | `he..o`   |        |        |
| `^`       | Starts with                         | `^hello`  |        |        |
| `$`       | Ends with                           | `planet$` |        |        |
| `*`       | Zero or more occurrences            | `he.*o`   |        |        |
| `+`       | One or more occurrences             | `he.+o`   |        |        |
| `?`       | Zero or one occurrence              | `he.?o`   |        |        |
| `{}`      | Exact number of occurrences         | `he.{2}o` |        |        |
| `         | `                                   | Either/or | `falls | stays` |
| `()`      | Capture and group                   | `(abc)`   |        |        |


## Flags

| Flag            | Shorthand | Description                  |
| --------------- | --------- | ---------------------------- |
| `re.ASCII`      | `re.A`    | Returns only ASCII matches   |
| `re.DEBUG`      |           | Returns debug information    |
| `re.DOTALL`     | `re.S`    | `.` matches newline too      |
| `re.IGNORECASE` | `re.I`    | Case-insensitive matching    |
| `re.MULTILINE`  | `re.M`    | Match start/end of each line |
| `re.NOFLAG`     |           | No flag specified            |
| `re.UNICODE`    | `re.U`    | Unicode matching             |
| `re.VERBOSE`    | `re.X`    | Allows comments and spaces   |


## Special Sequences

| Sequence | Description                    | Example    |
| -------- | ------------------------------ | ---------- |
| `\A`     | Match at beginning of string   | `\AThe`    |
| `\b`     | Match at beginning/end of word | `r"\bain"` |
| `\B`     | Match NOT at beginning/end     | `r"\Bain"` |
| `\d`     | Match digits                   | `\d`       |
| `\D`     | Match non-digits               | `\D`       |
| `\s`     | Match whitespace               | `\s`       |
| `\S`     | Match non-whitespace           | `\S`       |
| `\w`     | Match word characters          | `\w`       |
| `\W`     | Match non-word characters      | `\W`       |
| `\Z`     | Match at end of string         | `Spain\Z`  |


## Sets

| Set          | Description                              |
| ------------ | ---------------------------------------- |
| `[arn]`      | Match `a`, `r`, or `n`                   |
| `[a-n]`      | Match lowercase letters `a` to `n`       |
| `[^arn]`     | Match any character except `a`, `r`, `n` |
| `[0123]`     | Match `0`, `1`, `2`, or `3`              |
| `[0-9]`      | Match any digit                          |
| `[0-5][0-9]` | Match numbers from `00` to `59`          |
| `[a-zA-Z]`   | Match uppercase/lowercase letters        |
| `[+]`        | Match `+` symbol                         |

## Python Try Except
The try block lets you test a block of code for errors.

The except block lets you handle the error.

The else block lets you execute code when there is no error.

The finally block lets you execute code, regardless of the result of the try- and except blocks.

```python
try:
    print("Hello")

except:
    print("Error")

else:
    print("No error happened")
```

```python
try:
    print(x)

except:
    print("Error happened")

finally:
    print("Finished")
```
